# movie data_set

```python
Movie_details.head()
```

```python
Movie_details.info()
```

```python
Movie_details.isnull().sum()
```

```python
Movie_details['First_Actor']=Movie_details['actors'].str.split('|').str[0]
Movie_details['Second_Actor']=Movie_details['actors'].str.split('|').str[1]
Movie_details['Third_Actor']=Movie_details['actors'].str.split('|').str[2]
Movie_details['Fourth_Actor']=Movie_details['actors'].str.split('|').str[3]

# Add a new colums month extracting the month 'releaseDate' column
Movie_details['release_month']=Movie_details['releaseDate'].apply(lambda x: x.lstrip().split(' ')[1])
```

```python
Movie_details
```

```python
M=sns.countplot(data=Movie_details,x='genre',order=Movie_details['genre'].value_counts().index[0:20])
plt.xticks(rotation=90,fontsize=12)
plt.xlabel('Genre',fontsize=15,fontweight='bold')
plt.ylabel('Count of Total Movie',fontsize=15,fontweight='bold')
plt.title('Top 20 Movie Genre',fontsize=15,fontweight='bold')
```

```python
M=sns.countplot(data=Movie_details,x='releaseYear',palette='Paired')
plt.xticks(rotation=90,fontsize=12)
plt.xlabel('Release Year',fontsize=15,fontweight='bold')
plt.ylabel('Count of Total Movies',fontsize=15,fontweight='bold')
plt.title('Total of Movie Release each Year ',fontsize=15,fontweight='bold')
```

```python
g=sns.countplot(data=Movie_details,y='First_Actor',order=Movie_details['First_Actor'].value_counts().index[0:20],palette = 'inferno')
plt.xticks(rotation = 90, fontsize = 12)
plt.xlabel('Total Movies', fontsize = 15,fontweight = 'bold')
plt.ylabel('First Actor', fontsize = 15,fontweight = 'bold')
plt.title('Total number of Movies by First Actor', fontsize = 15,fontweight = 'bold')
```
